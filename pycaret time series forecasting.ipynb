```python
df=df[df['InvNet']>0]
```

```python
df['InvDate']=pd.to_datetime(df['InvDate'],errors='coerce')
```

```python
data=(df.groupby([pd.Grouper(key='InvDate',freq='MS')])['InvNet'].sum().reset_index())
```

```python
data=data.loc[(data['InvDate']>='2019-01-01')&(data['InvDate']<'2022-06-01')]
```
